```python
import pyspark
from delta import *

builder = pyspark.sql.SparkSession.builder.appName("MyApp") 
spark = configure_spark_with_delta_pip(builder)\
            .getOrCreate()

spark
# Verify by printing the Spark master
print(f"Spark is using master: {spark.conf.get('spark.master')}")
```

```python
# Sample DataFrame to test
data = [("Alice", 30), ("Bob", 45), ("Cathy", 29)]
columns = ["Name", "Age"]
df = spark.createDataFrame(data, schema=columns)

# Define Delta table path (within the container filesystem or mounted volume)
delta_path = "/mnt/delta-table"

```

```python
import os


try:
    os.makedirs(delta_path, exist_ok=True)  # Ensure the directory exists
    print(f"Directory '{delta_path}' is ready for use.")
except OSError as e:
    print(f"Error creating directory: {e}")
```

```python
df.show()
```

```python
# Write DataFrame to Delta format

df.write.format("csv").mode("overwrite").save(delta_path)

```

```python
# Read data back from Delta table
delta_df = spark.read.format("delta").load(delta_path)
print("Data read from Delta table:")
delta_df.show()
```
